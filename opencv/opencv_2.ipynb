## 图像阈值
- cv2.threshold(src, threhold, maxval, type)

```python
import cv2
```

```python
img = cv2.imread('dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
ret, thresh1 = cv2.threshold(img, 177, 255, cv2.THRESH_TOZERO)
ret, thresh2 = cv2.threshold(img, 177, 255, cv2.THRESH_TOZERO_INV)
ret, thresh3 = cv2.threshold(img, 177, 255, cv2.THRESH_BINARY)
ret, thresh4 = cv2.threshold(img, 177, 255, cv2.THRESH_BINARY_INV)
ret, thresh5 = cv2.threshold(img, 177, 255, cv2.THRESH_TRUNC)
```

```python
import matplotlib.pyplot as plt

plt.subplot(231), plt.imshow(thresh1), plt.title('tozero')
plt.subplot(232), plt.imshow(thresh2), plt.title('tozero_inv')
plt.subplot(233), plt.imshow(thresh3), plt.title('binary')
plt.subplot(234), plt.imshow(thresh4), plt.title('binay_inv')
plt.subplot(235), plt.imshow(thresh5), plt.title('trunc')
plt.subplot(236), plt.imshow(img), plt.title('original')

plt.show()
```

## 图像平滑

```python
img_noise = cv2.imread('img_noise.jpg')
plt.imshow(img_noise)
```

```python
# 均值滤波
# 简单的平均卷积操作，卷积核尺寸多为奇数
blur = cv2.blur(img_noise, (3, 3))
plt.imshow(blur)
```

```python
# 方框滤波
# 基本和均值滤波一样，可以选择是否归一化
# True情况下和均值滤波一样, -1表示结果图的颜色通道数和原图一致（一般都是-1）
box_1 = cv2.boxFilter(img_noise, -1, (3, 3), normalize=True)
box_2 = cv2.boxFilter(img_noise, -1, (2, 2), normalize=False)

plt.subplot(121), plt.imshow(box_1), plt.title('normalize')
plt.subplot(122), plt.imshow(box_2), plt.title('without_normalize')
```

```python
# 高斯滤波
# 卷积核中距离中心点近的权重大，距离远的权重小
gaussian = cv2.GaussianBlur(img_noise, (5, 5), 1)
plt.subplot(121), plt.imshow(gaussian), plt.title('gaussian')
plt.subplot(122), plt.imshow(img_noise), plt.title('original')
```

```python
# 中值滤波
# 用周围区域像素值的中间值来作为中心点平滑处理后的像素值
median = cv2.medianBlur(img_noise, 5)
plt.subplot(121), plt.imshow(median), plt.title('median')
plt.subplot(122), plt.imshow(img_noise), plt.title('original')
```
