## 图像形态学处理

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('dige.jpg')
plt.imshow(img), plt.title('original')
```

## 形态学--腐蚀操作
- cv2.erode(src, kernel, iterations)

```python
# 腐蚀操作相当于向内缩，变换后毛刺消失，笔画由粗变细
img = cv2.imread('dige.jpg')

kernel = np.ones((5, 5), np.uint8)
erosion = cv2.erode(img, kernel, iterations = 1)

plt.imshow(erosion)
```

```python
# 迭代次数的影响
kernel_small = np.ones((3, 3), np.uint8)
erosion_1 = cv2.erode(img, kernel_small, iterations = 1)
erosion_2 = cv2.erode(img, kernel_small, iterations = 2)
erosion_3 = cv2.erode(img, kernel_small, iterations = 3)

plt.subplot(131), plt.imshow(erosion_1), plt.title('erosion_1')
plt.subplot(132), plt.imshow(erosion_2), plt.title('erosion_2')
plt.subplot(133), plt.imshow(erosion_3), plt.title('erosion_3')
plt.show()
```

## 形态学——膨胀操作
- cv2.dilate(src, kernel, iterations)

```python
img = cv2.imread('dige.jpg')
kernel = np.ones((5, 5), np.uint8)
erosion = cv2.erode(img, kernel, iterations = 1)
plt.imshow(erosion), plt.title('erosion')
```

```python
img_dilate = cv2.dilate(erosion, kernel, iterations = 1)
plt.imshow(img_dilate), plt.title('dilate')
```

```python
dilate_2 = cv2.dilate(erosion, kernel, iterations = 2)
dilate_3 = cv2.dilate(erosion, kernel, iterations = 3)

plt.subplot(131), plt.imshow(img_dilate), plt.title('dilate_1')
plt.subplot(132), plt.imshow(dilate_2), plt.title('dilate_2')
plt.subplot(133), plt.imshow(dilate_3), plt.title('dilate_3')
plt.show()
```

## 开运算与闭运算
- cv2.morphology(src, type, kernel)
   - type: cv2_MORPH_OPEN：开运算；cv2_MORPH_CLOSE：闭运算

```python
# 开运算：先腐蚀再膨胀
img = cv2.imread('dige.jpg')

kernel = np.ones((5, 5), np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.subplot(121), plt.imshow(img), plt.title('original')
plt.subplot(122), plt.imshow(opening), plt.title('opening')
plt.show()
```

```python
# 闭运算：先膨胀再腐蚀
# 显然，闭运算无法消除毛刺
img = cv2.imread('dige.jpg')

kernel = np.ones((5, 5), np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('original')
plt.subplot(122), plt.imshow(closing), plt.title('closing')
plt.show()
```

## 梯度运算
- cv2.morphology(src, cv2_MORPH_GRADIENT, kernel)

```python
img = cv2.imread('pie.jpg')

kernel = np.ones((5, 5), np.uint8)
erosion = cv2.erode(img, kernel, iterations = 3)
dilate = cv2.dilate(img, kernel, iterations = 3)

res = np.hstack((img, erosion, dilate))
plt.imshow(res)
```

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
plt.imshow(gradient)
```

## 礼帽与黑帽
- 礼帽：原始输入－开运算
- 黑帽：闭运算－原始输入

```python
# 礼帽
img = cv2.imread('dige.jpg')

kernel = np.ones((5, 5), np.uint8)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
plt.imshow(tophat), plt.title('tophat')
```

```python
# 黑帽
img = cv2.imread('dige.jpg')

kernel = np.ones((5, 5), np.uint8)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
plt.imshow(blackhat), plt.title('blackhat')
```
