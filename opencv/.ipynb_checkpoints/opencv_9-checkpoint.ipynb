# Scale invariant Feature Transform(SIFT)
## 图像尺度空间
考虑图像在不同尺度下都存在的共同特征

尺度空间的获取常常是通过高斯模糊来实现的
![](./images_for_note/feature/sift_3.png)
![](./images_for_note/feature/sift_2.png)
> $\sigma$控制模糊程度；$\sigma$越大，相比原图变化越大，类似于标准差

![](./images_for_note/feature/sift_4.png)

![](./images_for_note/feature/sift_6.png)
### DOG空间极值检测
为了寻找尺度空间的极值点，每个像素点要与其图像域（同一尺度空间）以及尺度域（相邻尺度空间）的所有相邻点作比较，当其大于（或小于）所有相邻点时，该点就是极值点。

如下图所示，中间的检测到在其图像域中要与8个像素点进行比较，同时还要与尺度域中的18个点进行比较，因此一共需要进行26次比较。

> 当某一像素点所在图像域处于顶层或低层时，进行比较的像素点数量会减少
![](./images_for_note/feature/sift_7.png)
### 关键点的精确定位
上述极值检测得到的候选关键点是DOG空间的局部极值点，而且这些极值点均为离散的点，精确定位极值点的一种方法是对尺度空间DOG函数进行曲线拟合，计算拟合函数的极值点，从而实现关键点的精确定位。
![](./images_for_note/feature/sift_8.png)
![](./images_for_note/feature/sift_9.png)
> 相当于极值点校正

### 特征点的主方向
![](./images_for_note/feature/sift_11.png)
每个特征点可以得到三个信息（x, y, $\theta$, $\sigma$），即位置、方向和尺度。具有多个方向的特征点可以被复制成多份（例如某个次方向和主方向非常相近，常见比例为80%），然后将方向值分别赋给复制后的特征点，一个特征点就产生了多个坐标、尺度相等，但是方向不同的特征点。

### 生成特征描述
在完成关键点的梯度计算后，计算直方图统计邻域内像素的梯度和方向：
![](./images_for_note/feature/sift_12.png)
为了保证特征矢量的旋转不变性，要以特征点为中心，在附近邻域内将坐标轴旋转$\theta$角度，即将坐标轴旋转为特征点的主方向。
![](./images_for_note/feature/sift_13.png)
![](./images_for_note/feature/sift_14.png)
旋转之后的主方向为中心取8x8的窗口，求每个像素的梯度幅值和方向，箭头方向代表梯度方向，长度代表梯度幅值，然后利用高斯窗口对其进行加权运算，最后在每个4x4的小块上绘制8个方向的梯度直方图，计算每个梯度方向的累加值，即可形成一个种子点，即每个特征点由4个种子点组成，每个种子点有8个方向的向量信息。
![](./images_for_note/feature/sift_16.png)
> 论文中建议对每个关键点使用4x4共16个种子点来描述，这样一个关键点就会产生128维的SIFT特征向量（[4, 4, 8]）

![](./images_for_note/feature/sift_17.png)

### opencv sift函数使用

```python
# 导入库和图像
import cv2
import matplotlib.pyplot as plt
import numpy as np

# img = cv2.imread('lena.jpg') # lena
img = cv2.imread('./images_for_note/feature/test_1.jpg') # house
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
cv2.__version__ # opencv-python 3.4.3版本以上因为专利保护无法使用SIFT
```

### 得到特征点

```python
sift = cv2.xfeatures2d.SIFT_create()
kp = sift.detect(gray, None) # kp数据类型由opencv封装好，不易展示
```

```python
img = cv2.drawKeypoints(gray, kp, img) # 利用opencv给定函数显示关键点

plt.imshow(img, cmap='gray')
```

### 计算特征

```python
kp, des = sift.compute(gray, kp)
```

```python
print(np.array(kp).shape) # 特征点的数量
```
