## 视频数据读取
- cv2.VideoCapture可以捕获摄像头，利用数字来选择设备，例如0，1
- 如果读取本地视频文件，直接传入视频文件的路径即可

```python
vc = cv2.VideoCapture('test.mp4')
```

```python
# 检查是否正确打开
if vc.isOpened():
    open, frame = vc.read()
else:
    open = False
```

```python
while open:
    ret, frame = vc.read()
    if frame is None:
        break
    if ret is True:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('gray', gray)
        if cv2.waitKey(10) & 0xFF == 27:
            break
vc.release()
cv2.destroyAllWindows()
```

## 截取图像部分区域

```python
img = cv2.imread('curry.jpg')
part = img[0:200, 0:200]
cv_show('part', part)
```

## 提取颜色通道
- cv2.split

```python
# 提取蓝色
b,g,r = cv2.split(img)
cv_show('blue', b)
```

```python
# 提取红色
cur_img = img.copy()
cur_img[:, :, 0] = 0
cur_img[:, :, 1] = 0
cv_show('cur_img', cur_img)
```

## 颜色通道融合
- cv2.merge

```python
merge = cv2.merge((b,g,r))
cv_show('merge', merge)
```

## 边界填充
- cv2.copyMakeBorder

```python
img_dog = cv2.imread('dog.jpg')
```

```python
# up, down, left, right
top_size, down_size, _left_size, right_size = (50, 50, 50, 50)

replicate = cv2.copyMakeBorder(img_dog, top_size, down_size, _left_size, right_size, borderType=cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img_dog, top_size, down_size, _left_size, right_size, cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img_dog, top_size, down_size, _left_size, right_size, cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img_dog, top_size, down_size, _left_size, right_size, cv2.BORDER_WRAP)
constant = cv2.copyMakeBorder(img_dog, top_size, down_size, _left_size, right_size, cv2.BORDER_CONSTANT, value=0)
```

```python
# display the results
plt.subplot(231), plt.imshow(img_dog, 'gray'), plt.title('original')
plt.subplot(232), plt.imshow(replicate, 'gray'), plt.title('replicate')
plt.subplot(233), plt.imshow(reflect, 'gray'), plt.title('reflect')
plt.subplot(234), plt.imshow(reflect101, 'gray'), plt.title('reflect101')
plt.subplot(235), plt.imshow(wrap, 'gray'), plt.title('wrap')
plt.subplot(236), plt.imshow(constant, 'gray'), plt.title('constant')

plt.show()
```
